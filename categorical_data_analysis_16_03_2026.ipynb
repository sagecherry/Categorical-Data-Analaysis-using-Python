```python
import pandas as pd
import numpy as np
data={
    'Order_ID':['01','02','03','04','05','06','07','08','09'],
    'Category':['electronics','clothing','grocery','electronics','clothing','grocery','clothing','electronics','grocery'],
    'Payment_Method':['UPI','COD','UPI','COD','UPI','COD','UPI','COD','UPI'],
    'Delivery_type':['Standard','Express','Standard','Express','Standard','Express','Standard','Express','Standard'],
    'Customer_Type':['Regular','New','Regular','New','Regular','New','Regular','New','Regular']
}
df=pd.DataFrame(data)
print(df)
```

```python
#1Frequency Count
#count no. of orders in each category
print(df['Category'].value_counts())
```

```python
#2. Unique Val
#Display all unique categories
print(df['Category'].unique())
```

```python
#3.Number of Unique Values
print(df['Category'].nunique())
```

```python
#4. Frequency of Payment Method
print(df['Payment_Method'].value_counts())
```

```python
#5. Cross Tabulation
#Category vs Payment Method
print(pd.crosstab(df['Category'],df['Payment_Method']))
```

```python
#6. Cross tabulation of delivery vs customer type
print(pd.crosstab(df['Delivery_type'],df['Customer_Type']))
```

```python
#7. Percentage Distribution
print(df['Category'].value_counts(normalize=True)*100)
```

```python
#8. Filtering Data
electronics_orders=df[df['Category']=='electronics']
print(electronics_orders)
```

```python
#9. Grouping
grouped_data=df.groupby('Category')['Order_ID'].count()
print(grouped_data)
```

```python
#10.Sorting Data
sorted_data=df.sort_values(by='Category')
print(sorted_data)
```
